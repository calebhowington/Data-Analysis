```python
def plotS21(filename):
    x, y = loadtxt(filename, comments='!', skiprows=10, usecols=(0, 3), unpack=True)
    plt.plot(x*1e-9,y)
    plt.title = filename
    plt.xlabel = "Frequency (GHz)"
    plt.ylabel = "S21 (dB)"
```
