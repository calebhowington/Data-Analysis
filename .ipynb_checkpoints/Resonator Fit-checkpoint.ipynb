## Functions

```python
def lorentz(x,*p):
#     p = amp, f0, kappa
    return (real(p[0]/(1-1j*(2*pi*x - 2*pi*p[1])/(p[2]/2)))**2 + imag(p[0]/(1-1j*(2*pi*x - 2*pi*p[1])/(p[2]/2)))**2)**.5

def analyze(x,y):
#     Find Resonance
    f0 = x[where(y == y.max())][0]
    
# Convert db to Voltage
    yvolt = 10**(y/20.)
    fig=figure(
        title="Analysis of " + filename,
        x_axis_label="Frequency (GHz)",
        y_axis_label="Amplitude (V)")
    fig.circle(x*1e-9,yvolt, fill_color=None)
    p, res = curve_fit(lorentz, x, yvolt, [yvolt.max(),f0,.01])
    string = "f0 = " + str(round(p[1]*1e-9, 6)) + " GHz\n"
    string = string + "kappa = " + str(p[2])
    string = string + "Q = " + str(p[1]/(2*pi*p[2]))
    fig.line(x*1e-9,lorentz(x,*p), legend=string)
    print("f0 = " + str(p[1]*1e-9) + " GHz")
    print("kappa = " + str(p[2]))
    print("Q = " + str(p[1]/(2*pi*p[2])))
    show(fig)

def plotS21(filename):
    print(filename)
    x, y = loadtxt(filename, comments='!', skiprows=10, usecols=(0, 3), unpack=True)
    fig = figure(
        title = filename,
        x_axis_label="Frequency (GHz)",
        y_axis_label="|S21| (dBm)" )
    fig.circle(x,y, fill_color=None)
    show(fig)
    return x,y
```
